## Part 2: print() and f-strings

`print()` outputs text to the screen.
An **f-string** lets you embed variables directly inside text.

```python
print(f"My name is {name} and I am {age} years old.")
```

The `f` before the quote means "f-string". Variables inside `{}` are replaced with their values.

````python
name = "Alice"
mood = "happy"
intensity = 8

print(f"Hi, I'm {name}!")
print(f"Right now I feel {mood} — intensity {intensity}/10.")
````

## Part 3: if / elif / else — Making Decisions

Python makes decisions with `if`. **Indentation** (4 spaces) is required — it tells Python what code belongs inside the if.

```python
if mood == "happy":
    print("Play upbeat songs! 🎉")
elif mood == "sad":
    print("Play soothing songs. 🌧️")
else:
    print("Play chill songs. 🎵")
```

Note: `==` compares values; `=` assigns a value.

````python
mood = "happy"

if mood == "happy":
    print("Play upbeat songs! 🎉")
elif mood == "sad":
    print("Play soothing songs. 🌧️")
elif mood == "angry":
    print("Play calming music. 😤→😌")
else:
    print("Play chill background music. 🎵")
````
