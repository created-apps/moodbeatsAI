## Part 1: Functions — Reusable Blocks of Logic

A **function** is code you write once and call many times.

```python
def function_name(parameter):
    # code goes here (indented)
    return value
```

- `def` — declares the function
- `parameter` — a value the function receives
- `return` — sends a value back to the caller

````python
def greet(name):
    message = f"Hello, {name}! Welcome to MoodBeats."
    return message

result = greet("Alice")
print(result)
print(greet("Bob"))
````

### What happens without `return`?

If you forget `return`, the function returns `None` — an empty value.

````python
def greet_no_return(name):
    message = f"Hello, {name}!"
    # forgot to return!

result = greet_no_return("Alice")
print(result)   # prints: None
````

## Part 2: Lists — Ordered Collections

A **list** stores multiple values in order.

```python
songs = ["Happy", "Don't Stop Me Now", "Good Feeling"]
songs[0]   # → "Happy"   (indexing starts at 0)
songs[-1]  # → "Good Feeling"  (last item)
len(songs) # → 3
```

````python
songs = ["Happy", "Don't Stop Me Now", "Good Feeling", "Walking on Sunshine"]

print(songs[0])   # first
print(songs[-1])  # last
print(len(songs)) # count

# Loop through the list
for song in songs:
    print(f"🎵 {song}")
````

## Part 3: Dictionaries — Key-Value Maps

A **dictionary** maps keys to values — like a real dictionary maps words to definitions.

```python
theme = {
    "accent": "#6366f1",
    "emoji":  "😊",
    "label":  "Happy"
}
theme["emoji"]       # → "😊"
theme.get("bg", "#000")  # → "#000" (safe default if key missing)
```

**This is exactly how EMOTION_THEMES works in MoodBeats!**

````python
# A theme dict for the 'happy' emotion
happy_theme = {
    "accent": "#f59e0b",
    "bg":     "#fef3c7",
    "emoji":  "😊",
    "label":  "Happy"
}

print(happy_theme["emoji"])
print(happy_theme["accent"])
print(happy_theme.get("border", "#000000"))  # key doesn't exist → default
````

Build a simplified version of the real MoodBeats `EMOTION_THEMES` dict.

````python
EMOTION_THEMES = {
    "happy": {
        "accent": "#f59e0b",
        "bg":     "#fef3c7",
        "emoji":  "😊"
    },
    "sad": {
        "accent": "#3b82f6",
        "bg":     "#eff6ff",
        "emoji":  "😢"
    },
    # TODO: Add 'angry' theme (use red tones)
    # TODO: Add 'surprised' theme
    # TODO: Add 'neutral' theme (use grey tones)
}

def get_theme(emotion):
    """Return the theme dict for an emotion, or a default grey theme."""
    default = {"accent": "#6b7280", "bg": "#f9fafb", "emoji": "😐"}
    return EMOTION_THEMES.get(emotion, default)

# Test it
print(get_theme("happy"))
print(get_theme("sad"))
print(get_theme("unknown_emotion"))  # should return default
````
